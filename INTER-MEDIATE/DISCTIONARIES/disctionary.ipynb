# Dictionary in Python

A dictionary is a mutable, unordered collection of key-value pairs, where each key is unique and maps to a value.

## Important Points:
- Keys must be immutable (e.g., strings, numbers, tuples).
- Values can be of any type, including lists or other dictionaries.
- Access elements using keys: `dict[key]`.
- Methods like `keys()`, `values()`, `items()` for iteration.
- Dictionaries are not ordered in Python < 3.7; ordered in 3.7+.
- Use `in` to check for key existence.

```python
mydict = {"name": "Rudra", "age": 20, "city": "Vadodara",}
print(mydict)

#Accesing values of DICTIONARY 
print(mydict["name"])

#Appending key values to the dictionary
mydict["email"] = "mistryrudra@yahoo.com"
print("Disctionary after append",mydict)


#DELETE key value from DICTIONARY
del mydict["email"]
print("Disctionary after delete",mydict)

mydict.pop("age")
print("Disctionary after pop",mydict)

#Updating key value in DICTIONARY
mydict["age"] = 21
print("Disctionary after update",mydict)

#Loop through DICTIONARY
for key in mydict:
    print(key, ":", mydict[key])
```
