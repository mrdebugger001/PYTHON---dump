### Sets
- Unordered collection of unique elements.
- Mutable: add/remove elements; no duplicates.
- No indexing; use iteration or membership tests.
- Supports set operations: union (|), intersection (&), difference (-), symmetric difference (^).
- Creation: `{}` or `set()`; immutable: `frozenset()`.


```python
def add (a, b):
    return a + b

def add(a, b, c):
    return a + b + c

print(add(1, 2))
print(add(1, 2, 3))
```

```python

def modify_list(lst):
    lst.append(4)
my_list = [1, 2, 3]
modify_list(my_list)
print(my_list)  # Output: [1, 2, 3, 4]
```

```python
#wap for call by value and call by reference

def modify_value(x):
    x += 1
    return x

def modify_reference(lst):
    lst.append(4)
# Call by value
num = 5
modified_num = modify_value(num)
print("Original number:", num)  # Output: 5
print("Modified number:", modified_num)  # Output: 6
# Call by reference
my_list = [1, 2, 3]
modify_reference(my_list)
print("Modified list:", my_list)  # Output: [1, 2, 3, 4]
```
